```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def categorize_size(x: int):
    if 0 <= x <= 100:
        return '0 - 100'
    elif 100 <= x <= 500:
        return '100 - 500'
    elif 500 <= x <= 1000:
        return '500 - 1000'
    elif x >= 1000:
        return '1000+'
```

```python
def categorize_objects(x: int):
    """_summary_

    группы 0-5, 5-10, 10-30, 30-40, 40+ - сделать категории
    """
    if 0 <= x <= 5:
        return '0 - 5'
    elif 5 <= x <= 10:
        return '5 - 10'
    elif 10 <= x <= 30:
        return '10 - 30'
    elif 30 <= x <= 40:
        return '30 - 40'
    elif 40 <= x:
        return '40+'
```

```python
from mykprototypes import MyKPrototypes
from mykmodes import MyKModes
```

```python
df['ent_and_culture_objects_cat'] = df['ent_and_culture_objects'].apply(categorize_objects)
```

```python

K = range(1,10)
cost = []
for k in list(K):
    k_prototypes = MyKModes(clusters_amount=k,  runs_amount = 5, logging=False)
    k_prototypes.fit_predict(df[['location_type', 'city_population', 'pupils_amount', 'district', 'ent_and_culture_objects_cat']])
    cost.append(k_prototypes.cost_)

plt.plot(K, cost, 'x-')
plt.xlabel('No. of clusters')
plt.ylabel('Cost')
plt.title(f'Curve')
plt.show()
```

```python
k_prototypes = MyKModes(clusters_amount=4,  runs_amount = 10, logging=False)
clusters = k_prototypes.fit_predict(df[['location_type', 'city_population', 'pupils_amount', 'district', 'ent_and_culture_objects_cat']])
k_prototypes.cluster_centroids_
```
